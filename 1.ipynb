representativa expression value of gene cluster:



for simplicity, I' ll generate values randomly.





```python
import numpy as np

#X
p, n = 10 , 2
standardize = True
def mat(p, n):
    x = np.random.rand(p, n)
    print('original matrix:\n {}\n'.format(x))
    if standardize:
        means = np.mean(x)
        standard_deviation = np.std(x)
        if np.any(standard_deviation == 0):
            raise ValueError("There are predictor variables with st. dev. = 0")

        standardized_x = x - means / standard_deviation
        return standardized_x
    else:
        means = None
        standard_deviation = None
        return x
    
print('standardized matrix:\n {}\n'.format(mat(p, n)))
```
